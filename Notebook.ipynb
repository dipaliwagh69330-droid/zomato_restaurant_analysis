```python
df.groupby('restaurant type')['rate (out of 5)'].agg(['mean','count']).sort_values(by='mean',ascending = False).head(10)
```

```python
restaurant_rating_summary = df.groupby('restaurant type')['rate (out of 5)'].agg(['mean','count']).sort_values(by='mean',ascending = False).head(10)
```

```python
restaurant_rating_summary['mean'].plot(kind = 'barh')
plt.title('Top 10 restaurant types by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Restaurant Type')
plt.tight_layout()
plt.show()
```
